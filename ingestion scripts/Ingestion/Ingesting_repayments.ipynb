```python
from pyspark.sql import functions as F 
```

```python
csv_path = "/Volumes/bondara/raw/home_work_files/repayments.csv"
```

```python
target_table = "bondara.raw.ing_repayments"
```

```python
df_src = (
    spark.read.format("csv")
    .option("header", "true")
    .option("inferSchema", "true")
    .load(csv_path)
)
```

```python
df_src = df_src.withColumn("snapshot_ingested_at", F.current_timestamp())
```

```python
(
    df_src.write
    .format("delta")
    .mode("append")
    .option("overwriteSchema", "true")
    .saveAsTable(target_table)
)
```
