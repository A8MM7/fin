```python
from pyspark.sql import functions as F 
```

```python
csv_path = "/Volumes/bondara/raw/home_work_files/loan_product.csv"
```

```python
target_table = "bondara.raw.ing_loan_product_types"
```

```python
df = (
    spark.read.format("csv")
        .option("header", "true")
        .option("inferSchema", "true")
        .load(csv_path)
        .withColumn("snapshot_ingested_at", F.current_timestamp())
)

(
    df.write
        .format("delta")
        .mode("overwrite")
        .option("overwriteSchema", "true")
        .saveAsTable(target_table)
)
```
